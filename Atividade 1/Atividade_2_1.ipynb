## Explica√ß√£o das M√©tricas Estat√≠sticas

### 1. Coeficiente de Varia√ß√£o (CV)

**F√≥rmula:**

$$
CV = \left( \frac{\sigma}{\mu} \right) \times 100
$$

Onde:
- œÉ  = desvio padr√£o
- Œº  = m√©dia

**Interpreta√ß√£o:**
- Mede a **variabilidade relativa** dos dados em rela√ß√£o √† m√©dia.
- **CV < 15%**: indica baixa dispers√£o (dados homog√™neos).
- **CV > 30%**: indica alta dispers√£o (dados heterog√™neos).
- **15% <  CV < 30%**: indica m√©dia dispers√£o
- Usado para **comparar variabilidade** entre distribui√ß√µes com diferentes m√©dias.

---

### 2. Curtose

**F√≥rmula (Fisher=False):**

$$
\text{Curtose} = \frac{1}{n} \sum_{i=1}^{n} \left( \frac{x_i - \mu}{\sigma} \right)^4  - 3
$$

**Interpreta√ß√£o:**
- **Curtose > 0**: Distribui√ß√£o **leptoc√∫rtica** (pico alto, caudas pesadas).
- **Curtose = 0**: Distribui√ß√£o **mesoc√∫rtica** (normal).
- **Curtose < 0**: Distribui√ß√£o **platic√∫rtica** (achatada, caudas leves).
- Curtose elevada pode indicar **presen√ßa de outliers**.

---

### 3. Assimetria (Skewness)

**F√≥rmula:**

$$
\text{Skewness} = \frac{1}{n} \sum_{i=1}^{n} \left( \frac{x_i - \mu}{\sigma} \right)^3
$$

**Interpreta√ß√£o:**
- **Skewness ‚âà 0**: distribui√ß√£o sim√©trica.
- **Skewness > 0**: assimetria positiva (cauda √† direita).
- **Skewness < 0**: assimetria negativa (cauda √† esquerda).

---

### 4. M√©dia (Œº)

**F√≥rmula:**

$$
\mu = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

**Interpreta√ß√£o:**
- Representa o **valor m√©dio** dos dados.
- Sens√≠vel a valores extremos (**outliers**).
- Comparar com mediana e moda ajuda a entender a **simetria** da distribui√ß√£o.

---

### 5. Moda

**Defini√ß√£o:**

$$
\text{Moda} = \text{valor com maior frequ√™ncia no conjunto de dados}
$$

**Interpreta√ß√£o:**
- Valor mais comum.
- Pode haver:
  - **Uma moda** (unimodal)
  - **Duas modas** (bimodal)
  - **V√°rias modas** (multimodal)
- Diferen√ßas entre moda e m√©dia indicam **assimetria** da distribui√ß√£o.


```python
def calcular_metricas(coluna):
    cv = (coluna.std() / coluna.mean()) * 100      # Coeficiente de Varia√ß√£o (%)
    curtose = stats.kurtosis(coluna, fisher=False) # Curtose (Fisher=False: compara com normal)
    skewness = stats.skew(coluna)                  # Assimetria (Skewness)
    mean = coluna.mean()                           # M√©dia
    moda = coluna.mode().values[0]                 # Moda (pode haver m√∫ltiplas)

    print(f"Coeficiente de Varia√ß√£o (CV): {cv:.2f}%")
    print(f"Curtose (Fisher=False): {curtose:.2f}")
    print(f"Assimetria (Skewness): {skewness:.2f}")
    print(f"M√©dia: {mean:.2f}")
    print(f"Moda: {moda} \n")
```

## M√©tricas de Avalia√ß√£o de Modelos de Regress√£o

### 1. R-squared (R¬≤)

**O que √©:**  
Mede a propor√ß√£o da vari√¢ncia da vari√°vel dependente \( Y \) que √© explicada pelo modelo.

**F√≥rmula:**

$$
R^2 = 1 - \frac{\text{Vari√¢ncia n√£o explicada}}{\text{Vari√¢ncia total}}
$$

**Interpreta√ß√£o:**
- Varia de 0 a 1 (ou 0% a 100%).
- Exemplo: Se $R^2$ = 0,75 , ent√£o 75% da varia√ß√£o de \( Y \) √© explicada pelo modelo.

---

### 2. Adjusted R-squared (R¬≤ Ajustado)

**O que √©:**  
Vers√£o ajustada do R¬≤ que penaliza a inclus√£o de vari√°veis irrelevantes.

**F√≥rmula:**

$$
R^2_{\text{ajustado}} = 1 - \left( \frac{(1 - R^2)(n - 1)}{n - k - 1} \right)
$$

**Onde:**
- \( n \): n√∫mero de observa√ß√µes  
- \( k \): n√∫mero de vari√°veis independentes

**Interpreta√ß√£o:**
- Sempre menor ou igual ao R¬≤.
- Exemplo: Se \$R^2_{\text{ajustado}}$ = 0,72 enquanto R^2 = 0,75 , 3% da explica√ß√£o pode ser atribu√≠da a vari√°veis irrelevantes.

---

### 3. F-statistic

**O que √©:**  
Testa se pelo menos uma vari√°vel independente tem efeito significativo sobre \( Y \).

**F√≥rmula:**

$$
F = \frac{\text{Vari√¢ncia explicada}}{\text{Vari√¢ncia n√£o explicada}}
$$

**Interpreta√ß√£o:**
- Valores maiores indicam maior signific√¢ncia estat√≠stica do modelo.
- Exemplo: \( F = 80,12 \) sugere que o modelo √© altamente significativo.

---

```python
# Configura√ß√£o para reproducibilidade
np.random.seed(12)

# Vari√°vel X: m√©dia=10, desvio padr√£o=3, 30 pontos
x = np.random.normal(loc=10, scale=3, size=30)

# Vari√°vel Y: correlacionada com X (y = 2x + ru√≠do normal)
y = 0.8 * x + np.random.normal(loc=0.5, scale=2, size=30)  # loc=5 garante b positivo

# Valores de 'a' para testar
a_values = [-1.0, 0.0, 0.2, 0.4, 0.6, a_sk, 1.0, 1.2]
b_fixo = 0.23

resultados = []

for a in a_values:
    # Calcula y_pred para o 'a' atual
    y_pred = a * x + b_fixo

    # Calcula m√©tricas MANUALMENTE (sem OLS)
    residuos = y - y_pred
    n = len(y)
    k = 1  # n√∫mero de vari√°veis independentes

    # C√°lculo das estat√≠sticas
    ss_res = np.sum(residuos**2)
    ss_tot = np.sum((y - np.mean(y))**2)
    r2 = 1 - (ss_res / ss_tot)
    adj_r2 = 1 - ( (1-r2)*(n-1) / (n-k-1) )

    # F-statistic (simplificado)
    f_stat = ( (ss_tot - ss_res)/k ) / ( ss_res/(n-k-1) )

    # Armazena resultados
    resultados.append({
        'a': a,
        'R¬≤': r2,
        'R¬≤_adj': adj_r2,
        'F-statistic': f_stat,
        'SS_res': ss_res
    })

# Exibe tabela formatada
print(tabulate(resultados, headers="keys", tablefmt="github", floatfmt=".4f"))
```
